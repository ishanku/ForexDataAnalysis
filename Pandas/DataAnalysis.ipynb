```python
# Dependencies and Setup
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
import seaborn as sns
```

```python
forex_historical=pd.read_csv('../Resources/finalhistorical.csv')
forex_current=pd.read_csv('../Resources/2020_forex.csv')
```

```python
forex_historical=pd.DataFrame(forex_historical)
forex_current=pd.DataFrame(forex_current)
```

```python
fc=forex_current.drop(columns=['id']).set_index('Date')
```

```python
#fig, ((ax1, ax2, ax3, ax4 ), (ax5, ax6, ax7, ax8 ),(ax9, ax10, ax11, ax12 ),(ax13, ax14, ax15, ax16 ),(ax17, ax18, ax19, ax20 ),(ax21, ax22 )) = plt.subplots(nrows=5, ncols=4)
#fc.hist(bins=22, color='pink', edgecolor='black', linewidth=1.0,
#           xlabelsize=8, ylabelsize=10, grid=False)    
#plt.tight_layout(rect=(0, 0, 20, 10))
pp = sns.pairplot(fc, size=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))
fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Forex Current Currency Plots', fontsize=14)
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(20, 6))
corr = fc.corr()
htmap = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="Blues",fmt='.2f',linewidths=.05)
f.subplots_adjust(top=0.93)
Title= f.suptitle('Forex Current HeatMap', fontsize=14)
plt.savefig("../Graphs/ForexCurrentHeatMap.png")
```
