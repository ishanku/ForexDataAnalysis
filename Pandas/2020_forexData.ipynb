## Generate Dates List

```python
# List for holding 2020 dates which are missing from the foreign exchange datasets

sdate = date(2020, 1, 1)   # start date
edate = date.today()       # end date

delta = edate - sdate       # as timedelta
dates2020 = []

for i in range(delta.days + 1):
    day = sdate + timedelta(days=i)
    dates2020.append(day)
```

```python
#Practicing the data retrieval with One date

#For Specific Symbols
# https://data.fixer.io/api/2013-03-16?access_key=db56e416e100e59efb172517c271a05a&base=JPY&symbols=USD,AUD,CAD,PLN,MXN&format=1

# For All symbols
# https://data.fixer.io/api/2013-03-16?access_key=db56e416e100e59efb172517c271a05a&base=JPY&symbols=AUD,EUR,NZD,GPB,BRL,CAD,CNY,HKD,INR,KRW,MXN,ZAR,SGD,DKK,JPY,MYR,NOK,SEK&format=1

url = "https://data.fixer.io/api/"
access_key = forex_api_key
thedate = "2020-03-01"
base = "USD"
KaggelSymbols = "AUD,EUR,NZD,GPB,BRL,CAD,CNY,HKD,INR,KRW,MXN,ZAR,SGD,DKK,JPY,MYR,NOK,SEK,GBP,LKR,CHF,TWD,THB"

query_url = f"{url}{thedate}?access_key={forex_api_key}&base={base}&symbols={KaggelSymbols}&format=1"
response = requests.get(query_url).json()
pprint(response)

AUD = response['rates']['AUD']
EUR = response['rates']['EUR']
NZD = response['rates']['NZD']
BRL = response['rates']['BRL']
CAD = response['rates']['CAD']
CNY = response['rates']['CNY']
HKD = response['rates']['HKD']
INR = response['rates']['INR']
KRW = response['rates']['KRW']
MXN = response['rates']['MXN']
ZAR = response['rates']['ZAR']
SGD = response['rates']['ZAR']
DKK = response['rates']['DKK']
JPY = response['rates']['DKK']
MYR = response['rates']['MYR']
NOK = response['rates']['NOK']
SEK = response['rates']['SEK']
GBP = response['rates']['GBP']
LKR = response['rates']['LKR']
CHF = response['rates']['CHF']
TWD = response['rates']['TWD']
THB = response['rates']['THB']
```

### Perform API Calls
* Perform a forex retrieval for each date using a series of successive API calls.

```python
AUDList = []
EURList = []
NZDList = []
BRLList = []
CADList = []
CNYList = []
HKDList = []
INRList = []
KRWList = []
MXNList = []
ZARList = []
SGDList = []
DKKList = []
JPYList = []
MYRList = []
NOKList = []
SEKList = []
GBPList = []
LKRList = []
CHFList = []
TWDList = []
THBList = []

print ("Beginning the Data Retrieval")
print ('-'*50)
for dat in dates2020:
    try:
        query_url = f"{url}{dat}?access_key={forex_api_key}&base={base}&symbols={KaggelSymbols}&format=1"
        response = requests.get(query_url).json()
        AUDList.append(response['rates']['AUD'])
        EURList.append(response['rates']['EUR'])
        NZDList.append(response['rates']['NZD'])
        BRLList.append(response['rates']['BRL'])
        CADList.append(response['rates']['CAD'])
        CNYList.append(response['rates']['CNY'])
        HKDList.append(response['rates']['HKD'])
        INRList.append(response['rates']['INR'])
        KRWList.append(response['rates']['KRW'])
        MXNList.append(response['rates']['MXN'])
        ZARList.append(response['rates']['ZAR'])
        SGDList.append(response['rates']['SGD'])
        DKKList.append(response['rates']['DKK'])
        JPYList.append(response['rates']['JPY'])
        MYRList.append(response['rates']['MYR'])
        NOKList.append(response['rates']['NOK'])
        SEKList.append(response['rates']['SEK'])
        GBPList.append(response['rates']['GBP'])
        LKRList.append(response['rates']['LKR'])
        CHFList.append(response['rates']['CHF'])
        TWDList.append(response['rates']['TWD'])
        THBList.append(response['rates']['THB'])
    except:
        pass
print ("Data Retrieval Complete")
```

```python
forex_df = pd.DataFrame({"id": idList,
                        "Date": dates2020, 
                        "AUD": AUDList, 
                        "EUR": EURList, 
                        "NZD": NZDList, 
                        "BRL": BRLList, 
                        "CAD": CADList, 
                        "CNY": CNYList, 
                        "HKD": HKDList, 
                        "INR": INRList, 
                        "KRW": KRWList, 
                        "MXN": MXNList, 
                        "ZAR": ZARList, 
                        "SGD": SGDList, 
                        "DKK": DKKList, 
                        "JPY": JPYList, 
                        "MYR": MYRList, 
                        "NOK": NOKList, 
                        "SEK": SEKList, 
                        "GBP": GBPList, 
                        "LKR": LKRList, 
                        "CHF": CHFList, 
                        "TWD": TWDList, 
                        "THB":THBList})

forex_df.head()
```
